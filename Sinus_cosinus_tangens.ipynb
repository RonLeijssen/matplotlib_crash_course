### Load necessary libaries

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0,360.1,0.1)
y_sin = np.sin(np.deg2rad(np.arange(0,360.1,0.1)))
y_cos = np.cos(np.deg2rad(np.arange(0,360.1,0.1)))
```

```python
x_line=[0, 360.1]
y_line_up=[1, 1]
y_line_middle=[0, 0]
y_line_bottom=[-1, -1]
style = "--"
color = "green"
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot(x,y_sin,label='Sinus', color="red")
ax.plot(x,y_cos,label='Cosinus', color="blue")
ax.plot(x_line, y_line_up, linestyle=style, color=color)
ax.plot(x_line, y_line_middle, linestyle=style, color=color)
ax.plot(x_line, y_line_bottom, linestyle=style, color=color)

ax.set_xlabel('graden', fontsize=24)
ax.set_xticks(np.arange(0,370,10))
ax.set_yticks(np.arange(-1,1.1,0.1))

ax.grid(color="#dfdfdf")
ax.legend(fontsize=20, loc=(0.05,0.09))
plt.show()
```
