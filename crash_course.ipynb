```python
x = [0, 1, 2, 3]
y = [0, 2, 4, 6]
z = [0,4,8,12]
fontdict={'fontname':'comic sans ms','fontsize':24}

#fig, ax = plt.subplots(figsize=(16, 9))
#plt.subplots(figsize=(16, 9))
plt.figure(figsize=(16, 9), dpi=100)

plt.plot(x,y, label='2x', color='red', marker='.', markersize=10, linestyle='--')
plt.plot(x,z, label='4x', color='blue', marker='.', markersize=10)

plt.title('Onze Eerste Grafiek\n', fontdict)
plt.xlabel('x as', fontdict)
plt.ylabel('y as', fontdict)

#plt.xticks([0, 0.5, 0.25, 1, 2, 3, 4, 5])
plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks([0, 2 , 4, 6, 8])

plt.grid()
plt.legend(fontsize=18)
plt.savefig('mygraph.png', dpi=300)
plt.show()
```
